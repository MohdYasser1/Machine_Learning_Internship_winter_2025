# outliers check

```python
# Print All Boxplots
columns_Name = data.columns  # Get all column names from the DataFrame
for i in columns_Name:
    sns.boxplot(x=data[i], orient='h')  # Create a horizontal boxplot for each column
    plt.title(f'Boxplot of {i}')
    plt.show()  # Display the plot
```

```python
# Distribution Plots
for i in columns_Name:
    plt.figure()
    sns.histplot(data[i], kde=True)
    plt.title(f'Distribution of {i}')
    plt.show()
```

```python
# Pairplot for numerical columns
sns.pairplot(data.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#  العلاقة بين عدد الأطفال وعدد البالغين
plt.figure(figsize=(10, 6))
sns.countplot(x='number of adults', hue='number of children', data=data)
plt.title('Relationship between Number of Adults and Number of Children')
plt.show()
```

```python
#  توزيع الأسعار المتوسطة حسب نوع الغرفة
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='average price ', data=data)
plt.title('Distribution of Average Price by Room Type')
plt.show()
```

```python
# مقارنة عدد الحجوزات الملغاة مع الحجوزات غير الملغاة
plt.figure(figsize=(10, 6))
sns.countplot(x='booking status', data=data)
plt.title('Comparison of Canceled and Non-Canceled Bookings')
plt.show()
```
