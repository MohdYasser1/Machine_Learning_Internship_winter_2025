```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
```

```python
# Get dataset shape
num_rows, num_columns = df.shape

# Identify possible text and date/time features
text_features = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() > 50]
date_features = df.select_dtypes(include=['datetime64']).columns.tolist()

# Print results
print("Total Records (Rows):", num_rows)
print("Total Features (Columns):", num_columns)
print("Numerical Features:", len(numerical_cols))
print("Categorical Features:", len(categorical_cols))
print("Text Features:", text_features)
print("Date/Time Features:", date_features)
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Boxplots for outlier detection
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
def remove_outliers(df, numerical_cols):
    Q1 = df[numerical_cols].quantile(0.25)
    Q3 = df[numerical_cols].quantile(0.75)
    IQR = Q3 - Q1
    df_cleaned = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_cleaned

df = remove_outliers(df, numerical_cols)
```

```python
# Countplots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Countplot of {col}')
    plt.show()
```

```python
# Correlation heatmap (Only numerical features)
numeric_df = df[numerical_cols]
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
if len(numerical_cols) <= 5:
    sns.pairplot(df[numerical_cols])
    plt.show()
```

```python
# Display value counts for categorical columns
for col in categorical_cols:
    print(f"\nValue Counts for {col}:")
    print(df[col].value_counts())
```
